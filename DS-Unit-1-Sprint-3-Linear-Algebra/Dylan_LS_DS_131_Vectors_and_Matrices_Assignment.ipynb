## 1.1 Create a two-dimensional vector and plot it on a graph

```
# Import what I will need
import math
import matplotlib.pyplot as plt
import numpy as np

# Create my 2D vector
v_1 = [1, 1]

# Plot v_1, and make the graph
plt.arrow(0,0, v_1[0], v_1[1],head_width=.05, head_length=0.05, color ='blue')
plt.xlim(-1,2)          
plt.ylim(-1,2)
plt.title("A 2D Vector")
plt.show()
```

## 1.2 Create a three-dimensional vecor and plot it on a graph

```
# Import what I'll need
from mpl_toolkits.mplot3d import Axes3D

# Define a 3D vector
v_2 = [1, 1, 2]

# Plot it!
vectors = np.array([[0, 0, 0, 1, 1, 2]])

X, Y, Z, U, V, W = zip(*vectors)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=1)
ax.set_xlim([0, 2])
ax.set_ylim([0, 2])
ax.set_zlim([0, 2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
# Make them arrays
H = np.array([[12, 11], [7, 10]])
J = np.array([[0, 1, 2], [7, 10, 4], [3, 2, 0]])

print('Determinant of H:', np.linalg.det(H))
print('Determinant of J:', np.linalg.det(J))
```
